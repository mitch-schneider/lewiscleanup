## Lewis Core Permiability Consolidation

### Data Importation

```python
import pandas as pd

df = pd.read_csv('ALL_por_perm_data_CSV.csv')
df
```

### Defining Column Order of Ranked Importance

```python
col_order = ['K inf md',
           'Kair md', 
           'perm_md', 
           'perm_horz_md', 
           'Perm Ka Max md', 
           'perm_klink_md']
```

### Subsetting Dataframe 

```python
df2 = df[col_order]
df2
```

### Changing All Columns to Numeric

```python
df2['K inf md'] = pd.to_numeric(df2['K inf md'],errors='coerce')
df2['Kair md'] = pd.to_numeric(df2['Kair md'],errors='coerce')
df2['perm_md'] = pd.to_numeric(df2['perm_md'],errors='coerce')
df2['perm_horz_md'] = pd.to_numeric(df2['perm_horz_md'],errors='coerce')
df2['Perm Ka Max md'] = pd.to_numeric(df2['Perm Ka Max md'],errors='coerce')
df2['perm_klink_md'] = pd.to_numeric(df2['perm_klink_md'],errors='coerce')
```

### Creating Function to Pull Permeabilities from Dataframe in Ranked Order

```python
def singleval(row):
    if row['K inf md'] > 0:
        val = row['K inf md']
    elif row['Kair md'] > 0:
        val = row['Kair md']
    elif row['perm_md'] > 0:
        val = row['perm_md']
    elif row['perm_horz_md'] > 0:
        val = row['perm_horz_md']
    elif row['Perm Ka Max md'] > 0:
        val = row['Perm Ka Max md']
    else:
        row['perm_klink_md'] > 0
        val = row['perm_klink_md']
    return val

df2['final_perm'] = df2.apply(singleval, axis=1)
```
